```python
import pandas as pd
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/diogoalvesderesende/datasets/Datasets/Udemy_wikipedia_visits.csv',index_col='Date',parse_dates=True)
df.tail()
```

```python
df=df.asfreq("D")
```

```python
model=ExponentialSmoothing(endog=df.Udemy).fit()
```

```python
predictions=model.forecast(steps=31)
```

```python
df['Udemy']['2020-01-01':].plot()
predictions.plot()
```
