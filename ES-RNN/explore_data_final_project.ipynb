```python
import pandas as pd
df=pd.read_csv('run-of-river_production_load.csv',index_col='Date_Time',parse_dates=True)
```

```python
df=df.asfreq("D")
```

```python
df
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
model=ExponentialSmoothing(endog=df.Value).fit()
```

```python
predictions=model.forecast(steps=31)
```

```python
df['Value']['2020-01-01':].plot()
predictions.plot()
```
